```python
import pandas as pd
import numpy as np
import os
import sys
import random
```

```python
base_path = 'ifn_pictures'
```

```python
ifn_data = pd.read_csv('ifn_pictures/labels.csv')
ifn_data.head()
```

```python
#filter = ['Sunglasses', 'Shirt', 'Swimwear', 'Handbag', 'Highheels', 'Skirt']
filter = ifn_data.label.unique() # all labels, actually filter can be omitted 

booleans = []

#for every row in ifn_data check if label is in filter or not
for label in ifn_data.label:
    booleans.append(label in filter)

filter_pd = pd.Series(booleans)

ifn_data = ifn_data[filter_pd] #returns only those rows whose label is in filter
ifn_data
       
       
```

```python
# add a row to annotation file
def add_row_to_anno(idx,row, path, writer):   
    sys.stdout.write(str(idx) + '\r')
    sys.stdout.flush()
    
    x1 = int(row['xmin'])
    x2 = int(row['xmax'])
    y1 = int(row['ymin'])
    y2 = int(row['ymax'])
    
    img_path = os.path.join(path, row['image_path'])
    className = row['label']
    writer.write(img_path + ',' + str(x1) + ',' + str(y1) + ',' + str(x2) + ',' + str(y2) + ',' + className + '\n')    
```

```python
#create annotation file
path = '/content/drive/My Drive/CoLab/Faster R-CNN/data/IFN_Pics/'    
f= open(os.path.join(base_path, "annotation.txt"),"w+")
for idx, row in ifn_data.iterrows():
    add_row_to_anno(idx,row, path,f)    
f.close()
```

```python
#create separate test and training file

anno_test_name = "annotation_ifn_all_test.txt"
anno_train_name = "annotation_ifn_all_train.txt"
n_test_samples = 100 #numbers of test images

path = '/content/drive/My Drive/CoLab/Faster R-CNN/data/IFN_Pics/'


im_unique = ifn_data.image_path.unique()
random.shuffle(im_unique)
print('Total number of images: ', len(im_unique))

f= open(os.path.join(base_path, anno_test_name),"w+")
for i in range(n_test_samples):
    for _,row in ifn_data.iterrows():
        if row['image_path'] == im_unique[i]:
            add_row_to_anno(i,row, path,f)                            
f.close()    

f= open(os.path.join(base_path, anno_train_name),"w+")
for i in range(n_test_samples,len(im_unique)):
    for _,row in ifn_data.iterrows():
        if row['image_path'] == im_unique[i]:
            add_row_to_anno(i,row, path,f)                            
f.close()    


```
